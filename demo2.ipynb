```python
# Import the necessary libraries

from nn.network import NeuralNetwork
from nn.layer import Layer
from nn.loss import MSE, MSE_derivative
from nn.activations import tanh, tanh_derivative, sigmoid, sigmoid_derivative
from nn.gd import fit
from utils.loader import load_student_dataset

# data/student_grades/student-por.csv

import numpy as np
np.random.seed(8)

print("Lets form the neural network.")

nn = NeuralNetwork(loss=MSE, loss_derivative=MSE_derivative)

l2 = 0.001

layer1 = Layer(input_size=33, output_size=20, activation=tanh, d_activation=tanh_derivative, mask=None, l2_lambda=l2)
layer2 = Layer(input_size=20, output_size=10, activation=tanh, d_activation=tanh_derivative, mask=None, l2_lambda=l2)
layer3 = Layer(input_size=10, output_size=3, activation=sigmoid, d_activation=sigmoid_derivative, mask=None, l2_lambda=l2)

nn.add_layers([layer1, layer2, layer3])
```
